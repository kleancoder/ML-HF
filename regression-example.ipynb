```python
import numpy as np
import pandas as pd
```

```python
# load the data
# important node: this is where we will usually put data files
df = pd.read_csv('datasets/airfoil_self_noise.dat', sep='\t', header=None)
```

```python
# check the data
df.head()
df.info()
```

```python
# get the inputs
data = df[[0,1,2,3,4]].values
```

```python
data
```

```python
target = df[5].values
```

```python
target
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(data, target, test_size = 0.33)
```

```python
# instantiate a classifier and train it
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
# evaluate the model's performance
model.score(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
predictions = model.predict(X_test)
predictions
```

```python
# we can even use random forest to solve the same problem
from sklearn.ensemble import RandomForestRegressor
```

```python

model2 = RandomForestRegressor()
model2.fit(X_train, y_train)

```

```python
# evaluate model' performance
model2.score(X_train, y_train)
```

```python
model2.score(X_test, y_test)
```
